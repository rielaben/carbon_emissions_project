This function returns a series with year as the index and the sum of the column of interest by year as the only column

```python
def column_sums_by_year(df, col):
    return df.groupby(['year'])[col].sum()
```

The first visualization function in our python analysis plots time series data for a dataframe. This function is very flexible, as you can plot time series data (in this case, data by year) for one dataframe or a list of dataframes (both will be done in this analysis). It also allows you to pick the start year and end year they want to see plotted, assuming there is data available for those years.

```python
def time_series_plot(dfs, col, start_year=False, end_year = max_year, title='Global', max_yval=None, 
                     fill_below=True, list_countries=None):
    fig = plt.figure(figsize=(12, 5))
    list_country_index = 0
#     each df in 'dfs' resperesnts a dataframe for emissions data over time for 
#     a specific country
    for df in dfs:
#         returns year-by-year data for a particular column (such as 'co2 (M Tonnes)')
        df = column_sums_by_year(df, col)
#     if start year is not specified then the start year is the earliest year that data is available
        if start_year==False:
            start_year = min(df.index.year)

#     error checking to see if the inputted years have available data
        if (end_year > max(df.index.year))or(start_year < min(df.index.year)):
            print(f'ERROR: You must choose an end year between {min(df.index.year)} and {max(df.index.year)}')
            return
        
#     get data for year range
        series = df[str(start_year):str(end_year)]
    
#     extract x and y values and plot
        x_values = series.index
        y_values = series
        plt.plot(x_values, y_values)    

        if start_year == 1750:
            plt.title(f'{title} {col} emissions from the Industrial Revolution to {end_year}', fontsize=14)
        else:
            plt.title(f'{title} {col} emissions from {start_year} to {end_year}', fontsize=14)

        plt.xlabel('Year', fontsize=14)
        plt.ylabel(f'{col}', fontsize=14)
#         set y and x limits and shade below the line on graph
        if fill_below==True:
            plt.fill_between(x_values, y_values)
        plt.xlim([min(x_values), max(x_values)])
        if max_yval == None:
            plt.ylim([min(y_values), max(y_values)*1.05])
        else:
            plt.ylim([min(y_values), max_yval*1.05])
        
#         set legend if there are multiple countries
        if list_countries!=None:
            plt.legend(list_countries)
        
        plt.grid(True)
    return fig
```

```python
time_series_plot([country_data], 'co2 (M Tonnes)').savefig('imgs/global_co2_since_industrial_revolution.png')
time_series_plot([country_data], 'co2 (M Tonnes)', 1919).savefig('imgs/global_co2_1919_2019.png')
time_series_plot([country_data],'co2 (M Tonnes)',1950, 2000).savefig('imgs/global_co2_1950_2000.png')
```

This function returns data frame filtered by a specific year with countries sorted by descending values by the column of interest

```python
def country_data_by_year(df, year, col):
    return df[pd.DatetimeIndex(df['year']).year==year].sort_values(by=col, ascending=False)
```

The year_data_graph function returns a bar plot showing a ranking of the input column for a specific year. With this function, the user can input any year, column, and the number of countries they want to see in the bar plot.

```python
def year_data_graph(df, year, col, num_countries_display):
    data_year = country_data_by_year(df, year, col)
    
#     get median value, this will be plotted on the bar graph as well
    column_median_val = data_year[col].median()
    fig = plt.figure(figsize=(12, 5))
    
    plt.bar(data_year['country'][:num_countries_display].append(pd.Series(['Median All Countries'])),
            data_year[col][:num_countries_display].append(pd.Series([column_median_val])))
    plt.title(f'{year} Top {num_countries_display} Country {col} Levels')
    plt.xlabel('Country Name')
    plt.ylabel(f'{col}')
    plt.xticks(rotation = 90)
    plt.tight_layout()
    return fig
```

This function returns a number that sums up the input dataframe for the column of interest for the year of interest

```python
def get_col_value_for_year(df, year, col, country_list=None, opposite=False, total_df=False):
    data_by_year = country_data_by_year(df, year, col)
    
    if total_df==True:
        return data_by_year[col].sum()
    
#     This 'opposite' condition means to sum up the values for a specific column while excluding 
#     the countries that are in 'country list'. If opposite is False then it sums up
#     for the specified value with the filter being the countries in 'country list'
    if opposite == True:
        column_at_year_val = data_by_year
        for country in country_list:
            column_at_year_val = column_at_year_val[column_at_year_val['country'] != country]
        column_at_year_val = column_at_year_val[col].sum()
    else:
        column_at_year_val = data_by_year[data_by_year['country'].isin(country_list)]
    
        column_at_year_val = column_at_year_val[col].sum()
    return column_at_year_val
```

This function returns the ranked n counteis for a specific column

```python
def country_names_top_n_column(df, year, col, n_end, n_start_ = 0):
    return list(country_data_by_year(df, year, col)['country'][n_start_:n_end])

top_10_countries_co2_max_yr = country_names_top_n_column(country_data, max_year, 'co2 (M Tonnes)', 10)
top_10_countries_co2_max_yr
```

This function returns a dictionary where the keys are the top n countries for a specific year and a specific column and the values are what the values are for that column

```python
def dict_max_n_vals(df, year, col, n):
    df_dict = {}
    for country in country_names_top_n_column(df, year, col, n):
        country_n_df = get_col_value_for_year(df, year, col, [country])
        df_dict[country] = country_n_df

    return df_dict
```

The following code calculated various 'co2 (M Tonnes)' values for certain conditions using the functions defined above. We will be using these values to create pie charts in the functions below

```python
df_dict_c = dict_max_n_vals(country_data, max_year, 'co2 (M Tonnes)', 4)
display(df_dict_c)
df_curr_co2_keys = list(df_dict_c.keys())
df_curr_co2_values = list(df_dict_c.values())
print("keys: ", df_curr_co2_keys)
print("values: ", df_curr_co2_values)
display(df_curr_co2_values[0])


world_without_china_year_co2 = get_col_value_for_year(country_data, max_year, 'co2 (M Tonnes)', [df_curr_co2_keys[0]], opposite=True)
display(f"World minus {[df_curr_co2_keys[0]]} co2 (M Tonnes) in {max_year}", world_without_china_year_co2)

world_without_china_US_year_co2 = get_col_value_for_year(country_data, max_year, 'co2 (M Tonnes)', df_curr_co2_keys[:2], opposite=True)
display(f"World minus {df_curr_co2_keys[:2]} co2 (M Tonnes) in {max_year}", world_without_china_US_year_co2)

world_without_top_four_co2 = get_col_value_for_year(country_data, max_year, 'co2 (M Tonnes)', df_curr_co2_keys[:4], opposite=True)
display(f"World minus {df_curr_co2_keys[:4]} co2 (M Tonnes) in {max_year}", world_without_top_four_co2)

total_world_max_year = get_col_value_for_year(country_data, max_year, 'co2 (M Tonnes)', total_df=True)
display(f"Total world co2 (M Tonnes) in {max_year}", total_world_max_year)
```

```python
# checking to see if I calculated the totals correctly
np.testing.assert_almost_equal(df_curr_co2_values[0] + world_without_china_year_co2, total_world_max_year, 3)
np.testing.assert_almost_equal(world_without_china_US_year_co2 + df_curr_co2_values[0] + df_curr_co2_values[1], total_world_max_year, 3)
np.testing.assert_almost_equal(world_without_top_four_co2 + 
                               df_curr_co2_values[0] + df_curr_co2_values[1] + df_curr_co2_values[2] + df_curr_co2_values[3], total_world_max_year, 3)
```

This pie chart function takes in a list of values (which are calculated using the 'get_col_value_for_year' function) and creates a basic pie chart

```python
def pie_chart(input_nums, labels, title, total_val):
    
#     checking to see that the numbers to plot are correct
    total_val_num = sum(input_nums)
    np.testing.assert_almost_equal(total_val, total_val_num)
    
    zeros_list = []
    for i in input_nums:
        zeros_list.append(0)
    zeros_tuple = tuple(zeros_list)
    explode = zeros_tuple

    fig, ax = plt.subplots()
    ax.pie(input_nums, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
#     makes sure the elements are drawn in a circle
    ax.axis('equal')
    
    plt.title(title)
    return fig
    
pie_chart([world_without_china_year_co2, df_curr_co2_values[0]], ['Rest of World', 'China'], 
          f"China Co2 Emissions vs Rest of World for {max_year}", 
          total_world_max_year).savefig('imgs/china_vs_rest_world_pie_chart.png')
```

What does the relationship look like between the income of a country and the amount of co2 emissions produced?

The two functions below produces pie charts that visualize the percentage of co2 emitted by income category. You can change the column of interest to any column and the year to any year there is data on.

```python
def income_levels_pie_chart(df, col, year):
    display(list(df[col]))
    
    labels = list(df[col].index)
    sizes = list(df[col])
    explode = (0, 0, 0, 0)

    fig, ax1 = plt.subplots()
    ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    ax1.set_title(f"{col} By Income Level in {year}")

    return fig
```

```python
def co2_emissions_by_income(df, col, year):
    
    data_year = country_data_by_year(df, year, col)
#     graphing by sum for each income bracket
    sum_data = data_year.groupby('Income').sum()
    fig = income_levels_pie_chart(sum_data, col, year)
    return fig
    
co2_emissions_by_income(country_data, 'co2 (M Tonnes)', max_year).savefig('imgs/income_co2_pie_chart.png')
co2_emissions_by_income(country_data, 'cumulative_co2 (M Tonnes)', max_year).savefig('imgs/income_cumulative_co2_pie_chart.png')
```

At this point, it makes sense to start plotting correlations to see how strong the columns are different columns are related to each other. The best way to do this is by starting with a heat map and correlation matrix, and we can create this using the following function below. (https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e)

```python
def heat_map(df):
    corrmat = df.corr()
    top_corr_features = corrmat.index
    fig = plt.figure(figsize=(10,10))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    mask = np.triu(np.ones_like(corrmat))
    g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap=cmap, vmax=1, vmin=-1, mask=mask)
    return fig

heat_map(year_2018).savefig('imgs/heat_map.png')
```

The function below creates a scatter plot with two columns, and also displays the Pearson correlation as well as the p-value:

```python
def co2_emissions_scatter(df, x_col, y_col, title=None):
    test_data = df[(df[x_col].notna()) & (df[y_col].notna())]
    new_x = test_data[x_col]
    new_y = test_data[y_col]

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.scatter(x = new_x, y = new_y)
    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)

    test = stats.pearsonr(new_x, new_y)
    display("Test Statistics", test)
    
    return fig
```

However, there is one problem with this, and it is that countries differ in size in terms of population. There may be countries that, per capita, are worse polluters than the U.S., China, India, and some of the other major polluters, but because their population is so small we would have no idea by only looking at the analysis above. This is why we also need to account for these statistics per capita in addition to total emissions for entire countries.

This function below creates a paired t-test, breaking the data frame by desired income level and filtering by the column of interest.

```python
def paired_t_test(df, income_level, col_of_interest):
    cat1 = df[df['Income'] <= income_level]
    cat2 = df[df['Income'] > income_level]

    current_co2_per_capita_test = stats.ttest_ind(cat1[col_of_interest].dropna(), 
                                                  cat2[col_of_interest].dropna(),equal_var=False)
    print(f"{col_of_interest}: {current_co2_per_capita_test}")
    
data_max_year = country_data_by_year(country_data, 2019, 'co2_per_capita (Tonnes)')
paired_t_test(data_max_year, 'Lower middle', 'co2_per_capita (Tonnes)')
```

Running this code gives us a t-value of =-9.35, and a p-value=4.78e-17. The p-value indicated that the results are significant has a 5% threshold. The t-value is a ratio of the difference between two groups and the difference within the groups (the 2 groups being "low" and "lower middle" income countries vs. "upper middle" and "high income" countries). Essentially, the larger the t-score, the more difference there is between groups, and the smaller the t-score, the more similarity there is between groups. A magnitude higher than 9 is extremely high and indicates that there is a large difference between these two groups of incomes when it comes to co2 emissions per capita. (https://blog.minitab.com/en/statistics-and-quality-data-analysis/what-are-t-values-and-p-values-in-statistics)
